```python

import cv2
import numpy as np
import os
import tifffile

output_dir = "fourier_tif"
os.makedirs(output_dir, exist_ok=True)

fourier_magnitudes = []
fourier_magnitudesD = []

for i, image in enumerate(images[:100]):
    # Conversion en niveaux de gris si besoin
    if image.ndim == 3 and image.shape[2] == 3:
        gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    else:
        gray = image.astype(np.uint8)

    # Transformée de Fourier
    fourier = cv2.dft(np.float32(gray), flags=cv2.DFT_COMPLEX_OUTPUT)
    fourier_shift = np.fft.fftshift(fourier)
    magnitude = 20 * np.log(cv2.magnitude(fourier_shift[:, :, 0], fourier_shift[:, :, 1]) + 1e-5)
    magnitude_disp = cv2.normalize(magnitude, None, 0, 255, cv2.NORM_MINMAX).astype(np.uint8)
    fourier_magnitudesD.append(magnitude_disp)
    # Sauvegarde brute (non normalisée) pour analyse quantitative
    filename = f"fourier_{i:04d}.tif"
    filepath = os.path.join(output_dir, filename)
    tifffile.imwrite(filepath, magnitude.astype(np.float32))

    # On garde les magnitudes pour traitement
    fourier_magnitudes.append(magnitude)

```

```python
from scipy.optimize import curve_fit
import numpy as np
import matplotlib.pyplot as plt

# -- Paramètres --
frames = np.arange(start, end)
frame_rate = 20  # images par seconde
time_seconds = frames / frame_rate
f = 20275  # spatial frequency en m⁻¹
#30px 40551
#60px 20275
harmonics = [1, 2]
two_pi_sq_f2 = (2 * np.pi * f)**2

# -- Fonction à fitter (harmonique n en paramètre externe) --
def exp_model(t, A, D, C, n):
    tau_n = 1 / (two_pi_sq_f2 * n**2 * D)
    return A * np.exp(-t / tau_n) + C

# -- Fitting pour chaque harmonique --

fit_results = {}
for n, ydata in zip(harmonics, [first_norm, second_norm]):
    def model_to_fit(t, A, D, C):  # version réduite à 3 paramètres
        return exp_model(t, A, D, C, n)
   
    # Estimations initiales : A ~ 1, D ~ 10^-12, C ~ 0
    p0 = (1.0, 1e-12, 0.0)
    popt, pcov = curve_fit(model_to_fit, time_seconds, ydata, p0=p0, maxfev=10000)
    fit_curve = model_to_fit(time_seconds, *popt)

   
    A_fit, D_fit, C_fit = popt
    tau_fit = 1 / (two_pi_sq_f2 * n**2 * D_fit)

    fit_results[n] = {
        'D': D_fit,
        'tau': tau_fit,
        'A': A_fit,
        'C': C_fit,
        'fit_curve': fit_curve
    }

# -- Affichage des courbes et fits --
plt.figure(figsize=(10, 5))
plt.plot(time_seconds, first_norm, 'o', label="1st harmonic (data)")
plt.plot(time_seconds, fit_results[1]['fit_curve'], '-', label=f"1st fit (D={fit_results[1]['D']:.2e} m²/s)")

plt.plot(time_seconds, second_norm, 's', label="2nd harmonic (data)")
plt.plot(time_seconds, fit_results[2]['fit_curve'], '-', label=f"2nd fit (D={fit_results[2]['D']:.2e} m²/s)")

plt.xlabel("Time (s)")
plt.ylabel("Normalized FT intensity")
plt.title("Exponential fit of 1st and 2nd harmonics")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()






# -- Résultats numériques --
for n in harmonics:
    D_fit = fit_results[n]['D']
    tau_fit = fit_results[n]['tau']
    print(f"Harmonique n={n} : D = {D_fit:.3e} m²/s, τ = {tau_fit:.2f} s")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# -- Paramètres --
start = 21
end = 170
w = 2  # demi-largeur de la fenêtre d'intégration

# -- Frame de référence pour la détection --
ref_spectrum = fourier_magnitudes[start]
H, W = ref_spectrum.shape
cy, cx = H // 2, W // 2
profile = ref_spectrum[cy, :]  # Profil horizontal

# -- Détection automatique des pics --
peaks, _ = find_peaks(profile, height=profile.max()*0.1, distance=5)
peaks_right = peaks[peaks > cx]
relative_peaks = np.sort(peaks_right - cx)

if len(relative_peaks) < 3:
    raise RuntimeError("Moins de trois pics détectés dans le spectre.")

r1, r2, r3 = relative_peaks[:3]
print(f"Harmoniques détectées : r1={r1} px, r2={r2} px, r3={r3} px")

# -- Extraction temporelle des intensités --
first_trace = []
second_trace = []
third_trace = []

for i in range(start, end):
    F = fourier_magnitudes[i]

    zone1 = F[cy-w:cy+w+1, cx + r1 - w: cx + r1 + w + 1]
    zone2 = F[cy-w:cy+w+1, cx + r2 - w: cx + r2 + w + 1]
    zone3 = F[cy-w:cy+w+1, cx + r3 - w: cx + r3 + w + 1]

    first_trace.append(np.mean(zone1))
    second_trace.append(np.mean(zone2))
    third_trace.append(np.mean(zone3))

# -- Normalisation --
def normalize(trace):
    return (trace - np.min(trace)) / (np.max(trace) - np.min(trace))

first_norm = normalize(np.array(first_trace))
second_norm = normalize(np.array(second_trace))
third_norm = normalize(np.array(third_trace))

# -- Tracé --
frames = np.arange(start, end)

plt.figure(figsize=(10, 5))
plt.plot(frames, first_norm, 'o-', label="1st harmonic")
plt.plot(frames, second_norm, 's-', label="2nd harmonic")
plt.plot(frames, third_norm, '^-', label="3rd harmonic")
plt.xlabel("Frame index")
plt.ylabel("Normalized FT peak intensity [0–1]")
plt.title("Decay of 1st to 3rd harmonics (frames 21 to 65)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# -- Extraction harmonique 0 (pic central) --
w = 2  # demi-largeur de la zone
zeroth_trace = []

for i in range(start, end):
    F = fourier_magnitudes[i]
    zone0 = F[cy-w:cy+w+1, cx-w:cx+w+1]  # centre du spectre
    zeroth_trace.append(np.mean(zone0))

zeroth_trace = np.array(zeroth_trace)
zeroth_norm = (zeroth_trace - zeroth_trace.min()) / (zeroth_trace.max() - zeroth_trace.min())



r3, r4, r5 = sorted_peaks[2], sorted_peaks[3], sorted_peaks[4]

# Extraction des intensités pour chaque pic
fourth_trace = []
fifth_trace = []

for i in range(start, end):
    F = fourier_magnitudes[i]
    zone4 = F[cy-w:cy+w+1, cx + r4 - w: cx + r4 + w + 1]
    zone5 = F[cy-w:cy+w+1, cx + r5 - w: cx + r5 + w + 1]
    fourth_trace.append(np.mean(zone4))
    fifth_trace.append(np.mean(zone5))

fourth_trace = np.array(fourth_trace)
fifth_trace = np.array(fifth_trace)

fourth_norm = (fourth_trace - fourth_trace.min()) / (fourth_trace.max() - fourth_trace.min())
fifth_norm = (fifth_trace - fifth_trace.min()) / (fifth_trace.max() - fifth_trace.min())



# === Paramètres ===
frame_rate = 20  # images par seconde
f = 20275  # fréquence spatiale en m⁻¹ (correspond à 60 px)
harmonics = [0, 1, 2, 3,4,5]
harmonic_traces = [zeroth_norm, first_norm, second_norm, third_norm, fourth_norm, fifth_norm]

# === Temps en secondes ===
frames = np.arange(start, end)
time_seconds = frames / frame_rate

# === Modèle exponentiel simple ===
def simple_exp(t, A, tau, C):
    return A * np.exp(-t / tau) + C

# -- Paramètres dynamiques de fenêtre pour le fit --
min_duration = 2  # secondes minimum pour les hautes harmoniques
max_duration = (end - start) / frame_rate  # durée totale disponible

def get_fit_indices(n, time_array):
    """
    Renvoie les indices temporels pour le fit de l'harmonique n.
    Plus n est grand, plus la fenêtre est courte (proportionnelle à 1/n²).
    """
    total_duration = time_array[-1] - time_array[0]
    # Échelle inverse quadratique
    duration = max(min_duration, total_duration / (n**2)) if n > 0 else total_duration
    t_end = time_array[0] + duration
    indices = np.where((time_array >= time_array[0]) & (time_array <= t_end))[0]
    return indices


# === Fit exponentiel + extraction de D ===
fit_results = {}

for n, ydata in zip(harmonics, harmonic_traces):
    try:
        fit_idx = get_fit_indices(n, time_seconds)
        t_fit = time_seconds[fit_idx]
        y_fit = ydata[fit_idx]

        popt, pcov = curve_fit(simple_exp, t_fit, y_fit, p0=(1.0, 1.0, 0.0), maxfev=10000)
        A_fit, tau_fit, C_fit = popt
        sigma_tau = np.sqrt(pcov[1, 1])

        if n == 0:
            D_fit = None
            sigma_D = None
        else:
            f_term = (2 * np.pi * f)**2 * n**2
            D_fit = 1 / (f_term * tau_fit)
            sigma_D = (1 / f_term) * (sigma_tau / tau_fit**2)

        fit_curve = simple_exp(time_seconds, *popt)

        fit_results[n] = {
            'A': A_fit,
            'tau': tau_fit,
            'sigma_tau': sigma_tau,
            'C': C_fit,
            'D': D_fit,
            'sigma_D': sigma_D,
            'fit_curve': fit_curve,
            'success': True,
            'fit_indices': fit_idx
        }

    except RuntimeError:
        fit_results[n] = {'success': False}


# === Affichage des courbes ===
plt.figure(figsize=(12, 6))
colors = ['black', 'blue', 'green', 'orange', 'red', 'purple']

for i, n in enumerate(harmonics):
    y = harmonic_traces[i]
    label = "0ᵉ harmonic" if n == 0 else f"{n}ᵉ harmonic"
    plt.plot(time_seconds, y, 'o', color=colors[i], label=f"{label} (data)")
    if fit_results[n]['success']:
        plt.plot(time_seconds, fit_results[n]['fit_curve'], '-', color=colors[i],
                 label=f"{label} fit (τ={fit_results[n]['tau']:.2f}s" +
                       (f", D={fit_results[n]['D']:.2e} m²/s)" if n > 0 else ")"))

plt.xlabel("Time (s)")
plt.ylabel("Normalized FT intensity")
plt.title("Exponential decay fits of harmonics (n = 0 to 5)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Affichage numérique ===
for n in harmonics:
    if fit_results[n]['success']:
        print(f"Harmonique n={n} :")
        print(f"  τ = {fit_results[n]['tau']:.2f} ± {fit_results[n]['sigma_tau']:.2f} s")
        if n > 0:
            print(f"  D = ({fit_results[n]['D']:.3e} ± {fit_results[n]['sigma_D']:.1e}) m²/s")
    else:
        print(f"Harmonique n={n} : ❌ Fit échoué")

# === Fit quadratique de 1/tau = a * n² + b * n + c ===
n_vals_all = np.array(harmonics)
taus = np.array([fit_results[n]['tau'] if fit_results[n]['success'] else np.nan for n in harmonics])
tau_errs = np.array([fit_results[n]['sigma_tau'] if fit_results[n]['success'] else np.nan for n in harmonics])

valid = ~np.isnan(taus) & (n_vals_all > 0)  # exclude n=0 and failed fits

n_vals = n_vals_all[valid]
taus = taus[valid]
tau_errs = tau_errs[valid]
inv_taus = 1 / taus
inv_taus_err = tau_errs / taus**2

# === Tracé 1/τ vs. n ===
plt.figure(figsize=(8, 5))
plt.errorbar(n_vals, inv_taus, yerr=inv_taus_err, fmt='o', capsize=5, label="Data", color="darkgreen")

# === Fit quadratique : 1/τ = a·n² + b·n + c ===
def quadratic(n, a, b, c):
    return a * n**2 + b * n + c

popt, pcov = curve_fit(quadratic, n_vals, inv_taus, sigma=inv_taus_err, absolute_sigma=True)
a, b, c = popt
sigma_a = np.sqrt(pcov[0, 0])

n_fit = np.linspace(min(n_vals), max(n_vals), 100)
plt.plot(n_fit, quadratic(n_fit, *popt), '--', label=f"Fit: a·n² + b·n + c", color="gray")

plt.xlabel("Harmonic number n")
plt.ylabel("1 / τ (s⁻¹)")
plt.title("Decay rate vs Harmonic order")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Calcul D_eff à partir de a ===
D_eff = a / (2 * np.pi * f)**2
sigma_D_eff = sigma_a / (2 * np.pi * f)**2

print(f"\nFit quadratique: 1/τ = a·n² + b·n + c")
print(f"a = {a:.3e} ± {sigma_a:.1e} s⁻¹")
print(f"⇒ D_eff = ({D_eff:.3e} ± {sigma_D_eff:.1e}) m²/s")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# -- Paramètres --
frame_rate = 20  # images/s
f = 20275  # fréquence spatiale m⁻¹ (correspond à 60 px)
w = 2  # demi-largeur du carré autour du pic de Fourier

# === Extraction de l’harmonique 0 ===
zeroth_trace = []
for i in range(start, end):
    F = fourier_magnitudes[i]
    zone0 = F[cy-w:cy+w+1, cx-w:cx+w+1]
    zeroth_trace.append(np.mean(zone0))
zeroth_trace = np.array(zeroth_trace)

# === Extraction des harmoniques 1 à 5 (y compris 4 et 5 à partir des pics détectés) ===
r3, r4, r5 = sorted_peaks[2], sorted_peaks[3], sorted_peaks[4]

def extract_trace(r):
    trace = []
    for i in range(start, end):
        F = fourier_magnitudes[i]
        zone = F[cy-w:cy+w+1, cx + r - w: cx + r + w + 1]
        trace.append(np.mean(zone))
    return np.array(trace)

# Harmoniques 1 à 5 (déjà définis)
traces = [
    zeroth_trace,
    first_trace,
    second_trace,
    third_trace,
    extract_trace(r4),
    extract_trace(r5)
]

# === Correction : division par l’harmonique 0 frame par frame ===
epsilon = 1e-8
corrected_traces = [trace / (zeroth_trace + epsilon) for trace in traces]

# === Normalisation 0-1 de chaque trace corrigée ===
def normalize(x):
    xmin, xmax = np.min(x), np.max(x)
    if np.isclose(xmax, xmin):
        return np.zeros_like(x)  # ou np.ones_like(x), selon le sens physique
    return (x - xmin) / (xmax - xmin)

harmonic_traces = [normalize(trace) for trace in corrected_traces]


harmonic_traces = [normalize(trace) for trace in corrected_traces]

# === Paramètres temporels ===
frames = np.arange(start, end)
time_seconds = frames / frame_rate
harmonics = [0, 1, 2, 3]

# === Modèle exponentiel ===
def simple_exp(t, A, tau, C):
    return A * np.exp(-t / tau) + C

# === Fit avec fenêtres dynamiques ===
fit_results = {}
min_duration = 2  # durée min en secondes

def get_fit_indices(n, time_array):
    total_duration = time_array[-1] - time_array[0]
    duration = max(min_duration, total_duration / (n**2)) if n > 0 else total_duration
    t_end = time_array[0] + duration
    indices = np.where((time_array >= time_array[0]) & (time_array <= t_end))[0]
    return indices

for n, ydata in zip(harmonics, harmonic_traces):
    try:
        fit_idx = get_fit_indices(n, time_seconds)
        t_fit = time_seconds[fit_idx]
        y_fit = ydata[fit_idx]

        popt, pcov = curve_fit(simple_exp, t_fit, y_fit, p0=(1.0, 1.0, 0.0), maxfev=10000)
        A_fit, tau_fit, C_fit = popt
        sigma_tau = np.sqrt(pcov[1, 1])

        if n == 0:
            D_fit, sigma_D = None, None
        else:
            f_term = (2 * np.pi * f)**2 * n**2
            D_fit = 1 / (f_term * tau_fit)
            sigma_D = (1 / f_term) * (sigma_tau / tau_fit**2)

        fit_curve = simple_exp(time_seconds, *popt)

        fit_results[n] = {
            'A': A_fit,
            'tau': tau_fit,
            'sigma_tau': sigma_tau,
            'C': C_fit,
            'D': D_fit,
            'sigma_D': sigma_D,
            'fit_curve': fit_curve,
            'success': True,
            'fit_indices': fit_idx
        }

    except RuntimeError:
        fit_results[n] = {'success': False}

# === Tracé des courbes normalisées ===
colors = ['black', 'blue', 'green', 'orange', 'red', 'purple']
plt.figure(figsize=(12, 6))

for i, n in enumerate(harmonics):
    y = harmonic_traces[i]
    label = "0ᵉ harmonic" if n == 0 else f"{n}ᵉ harmonic"
    plt.plot(time_seconds, y, 'o', color=colors[i], label=f"{label} (data)")
    if fit_results[n]['success']:
        plt.plot(time_seconds, fit_results[n]['fit_curve'], '-', color=colors[i],
                 label=f"{label} fit (τ={fit_results[n]['tau']:.2f}s" +
                       (f", D={fit_results[n]['D']:.2e} m²/s)" if n > 0 else ")"))

plt.xlabel("Time (s)")
plt.ylabel("Normalized intensity (corrected by DC)")
plt.title("Exponential decay fits of harmonics (n = 0 to 5)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Résultats numériques ===
for n in harmonics:
    if fit_results[n]['success']:
        print(f"Harmonique n={n} :")
        print(f"  τ = {fit_results[n]['tau']:.2f} ± {fit_results[n]['sigma_tau']:.2f} s")
        if n > 0:
            print(f"  D = ({fit_results[n]['D']:.3e} ± {fit_results[n]['sigma_D']:.1e}) m²/s")
    else:
        print(f"Harmonique n={n} : ❌ Fit échoué")

# === Tracé et fit quadratique de 1/τ = a·n² + b·n + c ===
n_vals_all = np.array(harmonics)
taus = np.array([fit_results[n]['tau'] if fit_results[n]['success'] else np.nan for n in harmonics])
tau_errs = np.array([fit_results[n]['sigma_tau'] if fit_results[n]['success'] else np.nan for n in harmonics])

valid = ~np.isnan(taus) & (n_vals_all > 0)
n_vals = n_vals_all[valid]
taus = taus[valid]
tau_errs = tau_errs[valid]
inv_taus = 1 / taus
inv_taus_err = tau_errs / taus**2

plt.figure(figsize=(7, 5))
plt.errorbar(n_vals, inv_taus, yerr=inv_taus_err, fmt='o', capsize=5, color="darkgreen", label="Data")

def quadratic(n, a, b, c):
    return a * n**2 + b * n + c

popt, pcov = curve_fit(quadratic, n_vals, inv_taus, sigma=inv_taus_err, absolute_sigma=True)
a, b, c = popt
sigma_a = np.sqrt(pcov[0, 0])
n_fit = np.linspace(min(n_vals), max(n_vals), 200)
plt.plot(n_fit, quadratic(n_fit, *popt), '--', color='gray', label="Quadratic fit")

plt.xlabel("Harmonic number n")
plt.ylabel("1 / τ (s⁻¹)")
plt.title("Inverse decay time vs harmonic order")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# === Calcul de D_eff ===
D_eff = a / (2 * np.pi * f)**2
sigma_D_eff = sigma_a / (2 * np.pi * f)**2

print(f"\nFit quadratique: 1/τ = a·n² + b·n + c")
print(f"a = {a:.3e} ± {sigma_a:.1e} s⁻¹")
print(f"⇒ D_eff = ({D_eff:.3e} ± {sigma_D_eff:.1e}) m²/s")
```
